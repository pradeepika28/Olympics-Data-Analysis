```python
df.drop_duplicates(inplace=True)
df
```

```python
df['Medal'].fillna('No Medal', inplace=True) 
df['Height'].fillna(df['Height'].median(), inplace=True)
df['Weight'].fillna(df['Weight'].median(), inplace=True)
df['Age'].fillna(df['Age'].median(), inplace=True)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Age', 'Height', 'Weight']])
plt.title('Box plots for Age, Height, and Weight')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```

```python
Q1_age = df['Age'].quantile(0.25)
Q3_age = df['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

lower_age = Q1_age - 1.5 * IQR_age
upper_age = Q3_age + 1.5 * IQR_age

Q1_height = df['Height'].quantile(0.25)
Q3_height = df['Height'].quantile(0.75)
IQR_height = Q3_height - Q1_height

lower_height = Q1_height - 1.5 * IQR_height
upper_height = Q3_height + 1.5 * IQR_height

Q1_weight = df['Weight'].quantile(0.25)
Q3_weight = df['Weight'].quantile(0.75)
IQR_weight = Q3_weight - Q1_weight

lower_weight = Q1_weight - 1.5 * IQR_weight
upper_weight = Q3_weight + 1.5 * IQR_weight

outliers_age = df[(df['Age'] < lower_age) | (df['Age'] > upper_age)].index
outliers_height = df[(df['Height'] < lower_height) | (df['Height'] > upper_height)].index
outliers_weight = df[(df['Weight'] < lower_weight) | (df['Weight'] > upper_weight)].index

outliers = outliers_age.union(outliers_height).union(outliers_weight)

df_cleaned = df.drop(index=outliers)

print("New Shape after removing outliers: ", df_cleaned.shape)
df_cleaned
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_cleaned[['Age', 'Height', 'Weight']])
plt.title('Box plots for Age, Height, and Weight')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```

```python
medal_counts = df['Medal'].value_counts()
print(medal_counts)
```

```python
gold_medals = df[df['Medal'] == 'Gold']
gold_medal_teams = gold_medals.groupby('Team').size()
top_10_gold_medal_teams = gold_medal_teams.sort_values(ascending=False).head(10)
top_10_gold_medal_teams
```

```python
gold_medals = df[df['Medal'] == 'Silver']
gold_medal_teams = gold_medals.groupby('Team').size()
top_10_gold_medal_teams = gold_medal_teams.sort_values(ascending=False).head(10)
top_10_gold_medal_teams
```

```python
gold_medals = df[df['Medal'] == 'Bronze']
gold_medal_teams = gold_medals.groupby('Team').size()
top_10_gold_medal_teams = gold_medal_teams.sort_values(ascending=False).head(10)
top_10_gold_medal_teams
```

```python
medal_winners = df[df['Medal'] != 'No Medal']

team_medal_counts = medal_winners.groupby('Team').size()

top_10_teams = team_medal_counts.sort_values(ascending=False).head(10)

print(top_10_teams)
```

```python
team_medal_breakdown = medal_winners.groupby(['Team', 'Medal']).size().unstack(fill_value=0)
team_medal_breakdown['Total_Medals'] = team_medal_breakdown.sum(axis=1)
top_10_teams_with_breakdown = team_medal_breakdown.sort_values(by='Total_Medals', ascending=False).head(10)
top_10_teams_with_breakdown
```

```python
top_10_teams_with_breakdown[['Gold', 'Silver', 'Bronze']].plot(
    kind='bar', stacked=True, title="Top 10 Teams by Medal Type", color=['gold', 'silver', 'brown']
)
plt.xlabel("Team")
plt.ylabel("Medal Count")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Medal Type")
plt.tight_layout()
plt.show()
```

The United States is by far the most successful team, with over 1400 total medals, significantly outpacing the other top teams.

```python
athlete_medal_breakdown = medal_winners.groupby(['Name', 'Medal']).size().unstack(fill_value=0)
athlete_medal_breakdown['Total_Medals'] = athlete_medal_breakdown.sum(axis=1)
top_10_athletes_with_breakdown = athlete_medal_breakdown.sort_values(by='Total_Medals', ascending=False).head(10)
top_10_athletes_with_breakdown
```

```python
top_10_athletes_with_breakdown[['Gold', 'Silver', 'Bronze']].plot(
    kind='bar', stacked=True, title="Top 10 Athletes by Medal Type", color=['gold', 'silver', 'brown']
)
plt.xlabel("Athlete")
plt.ylabel("Medal Count")
plt.xticks(rotation=45, ha='right')
plt.legend(title="Medal Type")
plt.tight_layout()
plt.show()
```

Nikolay Yevmovich Andrianov has the highest total medal count among the top 10 athletes
Natalie Anne Coughlin (Hall) and Stefania Belmondo have a higher number of Gold medals than Nikolay Yevmovich Andrianov, despite Nikolay having the highest total medal count across all types.
